```python
import os, sys
import pandas as pd
```

```python
urls_path= os.getcwd() + '\\vtt_videos_urls.txt'
caps_path= os.getcwd() + '\\vtt_ground_truth.txt'
```

```python
f = open(caps_path, 'r')
captions = f.read().split("\n")
f.close()
captions
```

```python
captions[1]
```

```python
ids = []
caps = []
for i in captions[:-1]:
  id = i.split(' ', 1)[0]
  cap = i.split(' ', 1)[1]
  new_cap =  "<start> " + cap + " <end>"
  ids.append(id)
  caps.append(new_cap)
```

```python
df = pd.DataFrame()
df['captions'] =  caps 
df['id'] = ids
```

```python
df.head()
```

```python
feats_extract = []
# Loading all the numpy arrays at once and saving them in a dictionary
for filename in os.listdir(os.getcwd() + '\\features'):
    feats_extract.append(int(filename[:-4]))
```

```python
feats_extract
```

```python
df['id'] = df['id'].astype(int)
```

```python
df = df[df.id.isin(feats_extract)]
```

```python
df.head()
df.to_csv(os.getcwd() + '\\captions_lists\\train_captions.csv', index=False)
```

```python
train_list = df.values.tolist()
```

```python
import random
train_val_split = 0.85
random.shuffle(train_list)
training_list = train_list[:int(len(train_list)*train_val_split)]
validation_list = train_list[int(len(train_list)*train_val_split):]
```

```python
df = pd.DataFrame(training_list)
df.to_csv(os.getcwd() + '\\captions_lists\\training_list.csv', index=False)
df = pd.DataFrame(validation_list)
df.to_csv(os.getcwd() + '\\captions_lists\\validation_list.csv', index=False)
```
